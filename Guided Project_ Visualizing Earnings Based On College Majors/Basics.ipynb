We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.

We will try to answer the following questions in this project:<br>
1.Do students in more popular majors make more money?<br>
2.How many majors are predominantly male? Predominantly female?<br>
3.Which category of majors have the most students?
